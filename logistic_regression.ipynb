```python
import warnings
warnings.filterwarnings("ignore") #To avoid warnings.
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
logdf = pd.read_csv("insurance.csv")
lindf = pd.read_csv("insurance.csv")
```

```python
plt.scatter(lindf['age'],lindf['bought_insurance'],marker='*',color='blue')
```

```python
x = logdf[['age']]
y = logdf[['bought_insurance']]
```

```python
x
```

```python
y
```

```python
model = LogisticRegression()
```

```python
model.fit(x,y)
```

```python
age = int(input())
ans = model.predict([[age]])
ans[0]
```

```python
model.score(x,y)
```

```python
#train and split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.17)
model1 = LogisticRegression()
model1.fit(x_train,y_train)
```

```python
age = int(input())
ans = model1.predict([[age]])
ans[0]
```

```python
model1.score(x_train,y_train)
```
